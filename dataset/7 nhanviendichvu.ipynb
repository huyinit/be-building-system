```python
import random

# Danh sách mức lương cố định
muc_luong_co_dinh = [11000000, 12000000, 13000000, 14000000, 15000000, 9000000, 8000000]

# Tạo danh sách để lưu dữ liệu NhanVienDichVu
nhan_vien_dich_vu_docs = []

# Lặp qua từng nhân viên để tạo dữ liệu ngẫu nhiên
for nhan_vien in nhan_viens:
    for year in range(2020, 2025):  # Năm từ 2020 đến 2024
        for month in range(1, 13):  # Tháng từ 1 đến 12
            # Chọn ngẫu nhiên lương cho nhân viên từ danh sách mức lương cố định
            luong = random.choice(muc_luong_co_dinh)

            # Chọn ngẫu nhiên 1 đến 3 dịch vụ mà nhân viên làm
            num_services_used = random.randint(1, 3)
            services_used = random.sample(services, num_services_used)

            danh_sach_dich_vu = [service['_id'] for service in services_used]

            # Tạo tài liệu cho NhanVienDichVu
            nhan_vien_dich_vu_data = {
                'nhanVienId': nhan_vien['_id'],  # ID của nhân viên
                'nam': year,  # Năm làm việc
                'thang': month,  # Tháng làm việc
                'luong': luong,  # Lương của nhân viên
                'danhSachDichVu': danh_sach_dich_vu  # Danh sách dịch vụ mà nhân viên làm
            }

            # Thêm tài liệu vào danh sách
            nhan_vien_dich_vu_docs.append(nhan_vien_dich_vu_data)

# Chèn tất cả tài liệu vào collection NhanVienDichVu bằng insert_many
if nhan_vien_dich_vu_docs:
    print(nhan_vien_dich_vu_docs)
    nhan_vien_dich_vu_collection.insert_many(nhan_vien_dich_vu_docs)

print("Dữ liệu ngẫu nhiên cho NhanVienDichVu đã được tạo thành công.")
```

```python
import json

# Gửi lệnh ping để xác nhận kết nối thành công
try:
    client.admin.command('ping')
    print("Pinged your deployment. You successfully connected to MongoDB!")
    
    # Chọn cơ sở dữ liệu và collection
    db = client['building']  # Thay bằng tên cơ sở dữ liệu của bạn
    nhan_vien_dich_vu_collection = db['NhanVienDichVu']
    
    # Lấy tất cả dữ liệu từ collection
    nhanviendichvus = list(nhan_vien_dich_vu_collection.find())
    
    # Chuyển đổi ObjectId sang string để lưu vào JSON
    for nhanviendichvu in nhanviendichvus:
        nhanviendichvu['_id'] = str(nhanviendichvu['_id'])
        nhanviendichvu['nhanVienId'] = str(nhanviendichvu['nhanVienId'])
        
        for i in range(len(nhanviendichvu['danhSachDichVu'])):
            nhanviendichvu['danhSachDichVu'][i] = str(nhanviendichvu['danhSachDichVu'][i])

        print(nhanviendichvu)
    
    # Lưu dữ liệu vào file JSON
    with open('nhanviendichvu.json', 'w') as file:
        json.dump(nhanviendichvus, file, indent=4, ensure_ascii=False)
    
    print("Data has been saved to nhanviendichvu.json")
    
except Exception as e:
    print(e)
```
